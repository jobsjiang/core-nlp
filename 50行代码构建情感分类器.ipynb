```python
with open("../dataset/sentiment labelled sentences/amazon_cells_labelled.txt") as f:
    lines = f.readlines()
with open("../dataset/sentiment labelled sentences/imdb_labelled.txt") as f:
    temp = f.readlines()
    lines = lines + temp
with open("../dataset/sentiment labelled sentences/yelp_labelled.txt") as f:
    temp = f.readlines()
    lines = lines + temp
```

```python
x = []
y = []
for value in lines:
    temp = value.split('\t')
    x.append(temp[0])
    temp[1].replace('\n','')
    y.append(int(temp[1]))
```

```python
from keras.preprocessing.text import Tokenizer
```

```python
# Tokenizer实际功能
tokenizer_text = Tokenizer()
texts = ["The sun is shining in June!","September is grey.","Life is beautiful in August.","I like it","This and other things?"]
tokenizer_text.fit_on_texts(texts)
print(tokenizer_text.word_index)
```

```python
tokenizer_text.texts_to_sequences(["June is beautiful and I like it!"])
```

```python
tokenizer = Tokenizer(num_words=2500,split=' ')
tokenizer.fit_on_texts(x)
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
x
```

```python
X = tokenizer.texts_to_sequences(x)
X = pad_sequences(X) # pad_sequences会找出最长的句子，并用0填充其他较短语句以匹配该长度
```

```python
pad_sequences([[1, 2, 3], [3, 4, 5, 6], [7, 8]])
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# 将标签转换为one-hot编码，这有助于LSTM网络预测文本的标签
Y = []
for val in y:
    if(val == 0):
        Y.append([1,0])
    else:
        Y.append([0,1])
Y = np.array(Y)
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

```python
import keras
from keras.layers import Embedding,LSTM,Dense
from keras.models import Sequential
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Embedding(2500,128,input_length=X.shape[1],dropout=0.2))
model.add(LSTM(300,dropout_U=0.2,dropout_W=0.2))
model.add(Dense(2,activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,optimizer='adam',metrics=['accuracy'])
model.fit(X_train,y_train,epochs=10,verbose=2,batch_size=32)
print(model.evaluate(X_test,y_test)[1])
```
